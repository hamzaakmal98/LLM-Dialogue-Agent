```python

import torch
import numpy as np
from fastapi import FastAPI
from pydantic import BaseModel
from pyngrok import ngrok
from threading import Thread
from transformers import AutoTokenizer, AutoModelForCausalLM
import nest_asyncio
import json


# Check if GPU (CUDA/MPS) is available
if torch.cuda.is_available():
    device = torch.device('cuda')
    print("CUDA available! Running on GPU.", flush=True)
elif torch.backends.mps.is_available():
    device = torch.device('mps')
    print("MPS available! Running on GPU.", flush=True)
else:
    device = torch.device('cpu')
    print("CUDA NOT available... Running on CPU.", flush=True)

# Load Model and Tokenizer
access_token = "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"  # Replace with your actual token

# model_checkpoint = "gpt2-xl"
# tokenizer = AutoTokenizer.from_pretrained(model_checkpoint, token= access_token)
# model = AutoModelForCausalLM.from_pretrained(model_checkpoint, torch_dtype=torch.bfloat16, token=access_token)

prompt = ("Your role is to ensure that our AI-powered chatbot, designed to assist small businesses in creating automated marketing content, "
          "complies with all relevant laws and regulations. This includes data protection and privacy laws, AI ethics guidelines, and any regulations "
          "specific to marketing services. You should also conduct a risk assessment to identify any potential legal risks associated with the product. "
          "The deliverables should include a comprehensive legal compliance report and a risk mitigation plan. Give a detailed risk mitigation plan on "
          "how to combat these issues in the current day and age. DO NOT REPEAT THIS PROMPT IN YOUR OUTPUT.")


tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B-Instruct")
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-3.2-1B-Instruct")

model.to(device)  # Move model to appropriate device


```

```python

app = FastAPI()
class ProjectRequest(BaseModel):
  task: str

@app.post('/generate')
async def generate_text(request: ProjectRequest):

    print(f'API HIT!!')

    try:
        data = prompt + ". Do all the above for this project: " + request.task

        if not prompt:
            return json.dumps({"error": "No prompt provided"}), 400

        # Tokenize input and move to the appropriate device
        inputs = tokenizer(data, return_tensors="pt").to(device)

        # Generate text with specified parameters
        outputs = model.generate(
            inputs.input_ids,
            max_length=1024,
            min_length=500,
            num_beams=5,
            no_repeat_ngram_size=3,
            early_stopping=True
        )

        output_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

        print(f'Output Text : \n {output_text}')
        response_start_index = len(data)
        response = output_text[response_start_index:].strip()
        #response = output_text

        print(f'Final Response : \n {type(response)}')
        final = {"response_text": response}
        print(type(final))
        return final

    except Exception as e:
        return json.dumps({"error": str(e)}), 500

from pyngrok import ngrok

# Set your ngrok authtoken
ngrok.set_auth_token("2tPtZ9OKATsdkx3rwhYmGMgBOOE_5RcvfHT8QQxdgDi3wTt2U")

public_url = ngrok.connect(8000).public_url
print(f"Public API URL: {public_url}")

# # Run the Flask app
# app.run(port=5000)

nest_asyncio.apply()

import uvicorn

public_url = ngrok.connect(8000).public_url
uvicorn.run(app,host="0.0.0.0", port =8000)
```
